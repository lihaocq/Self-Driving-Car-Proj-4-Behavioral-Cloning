```python
############################
# 1. Get the image data and measurements
############################

import csv
import cv2
import numpy as np
from scipy import ndimage

##### read the .csv file for image path and steering measurements
lines =  []
with open('data/driving_log.csv') as csvfile:
    reader  = csv.reader(csvfile)
    firstline = True  # skip the first line
    for line in reader:
        if firstline == True:
            firstline = False
            continue
        lines.append(line)

images = []
measurements = []
correction = 0.2

##### get images of there 
for line in lines:
    for i in range(3):
        source_path = line[i+0]
        filename = source_path.split('/')[-1]
        current_path = 'data/IMG/' + filename
        image = ndimage.imread(current_path)
#       image = cv2.imread(current_path)    # cv2.imread will get images in BGR format, while drive.py uses RGB. So ndimage is used here.
        images.append(image)
        
        if i == 1:
            measurement = float(line[i+3]) + correction
        elif i==2:
             measurement = float(line[i+3]) - correction
        else:
            measurement = float(line[i+3])
        
        measurements.append(measurement)
```

```python
##############################
# Cropping image
#############################

import matplotlib.pyplot as plt

print('number of images',len(images))
img = images[0]
print('shape of image',img.shape)

fig, axs = plt.subplots(1, 2, figsize=(18, 9))
crop_img = img[60:(160-35),:,:]
axs[0].imshow(img)
axs[0].set_title('Before Cropping')
axs[1].imshow(crop_img)
axs[1].set_title('After Cropping')
```
